```python
from dotenv import load_dotenv
import os
import requests
import json
import pandas as pd 
import csv
import time
load_dotenv('API_key.env')
API_KEY=os.getenv('TMDB_API_KEY')
if API_KEY:
    print("API Key is Successfully loaded")
else: 
    print("API Key is not loaded")
```

```python
def get_movie_details(movie_name):
    # The TMDb API search endpoint
    search_url = f'https://api.themoviedb.org/3/search/movie?api_key={API_KEY}&query={movie_name}&language=en-US&page=1'

    # Make the request to the API
    response = requests.get(search_url)

    if response.status_code == 200:
        data = response.json()
        if data['results']:
            # Get the first movie result
            movie = data['results'][0]
            print(f"Title: {movie['title']}")
            print(f"Release Date: {movie['release_date']}")
            print(f"Overview: {movie['overview']}")
            print(f"Vote Average: {movie['vote_average']}")
            print(f"Popularity: {movie['popularity']}")
        else:
            print("No movie found with that name.")
    else:
        print(f"Failed to retrieve movie details. Status code: {response.status_code}")
# Get movie name input from user
movie_name = "The Dark Knight"

# Call the function to display movie details
get_movie_details(movie_name)
```

```python
def get_all_movies():
    # Open the CSV file for writing
    with open('all_movie_details.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=[
            'movie_id', 'title', 'overview', 'genres', 'release_date', 'runtime', 'cast', 'crew', 
            'original_language', 'poster_path', 'vote_average', 'vote_count', 'popularity', 
            'production_companies', 'keywords'
        ])
        
        writer.writeheader()

        page = 1
        while True:
            # The TMDb API popular movie endpoint
            url = f'https://api.themoviedb.org/3/movie/popular?api_key={API_KEY}&language=en-US&page={page}'
            response = requests.get(url)

            if response.status_code == 200:
                data = response.json()
                if data['results']:
                    # Loop through the results and write each movie's details to the CSV file
                    for movie in data['results']:
                        movie_id = movie['id']
                        movie_info = get_movie_details(movie_id)

                        if movie_info:
                            writer.writerow(movie_info)
                            print(f"Written details for movie: {movie_info['title']}")

                    page += 1
                    time.sleep(1)  # To avoid hitting rate limits
                else:
                    # No more movies available (end of data)
                    break
            else:
                print(f"Failed to retrieve data on page {page}. Status code: {response.status_code}")
                break

    print(f"All movie details have been written to 'all_movie_details.csv'")

# Call the function to get all movies
get_all_movies()
```

```python
def get_movie_details(movie_id):
    # Fetch the movie details from TMDb API
    url = f'https://api.themoviedb.org/3/movie/{movie_id}?api_key={API_KEY}&language=en-US'
    response = requests.get(url)
    
    if response.status_code == 200:
        movie = response.json()
        
        # Collect the required columns from the movie data
        movie_info = {
            'movie_id': movie.get('id', ''),
            'title': movie.get('title', ''),
            'overview': movie.get('overview', ''),
            'genres': ', '.join([genre['name'] for genre in movie.get('genres', [])]),  # List of genre names
            'release_date': movie.get('release_date', ''),
            'runtime': movie.get('runtime', ''),
            'original_language': movie.get('original_language', ''),
            'poster_path': f"https://image.tmdb.org/t/p/w500{movie.get('poster_path', '')}",
            'vote_average': movie.get('vote_average', ''),
            'vote_count': movie.get('vote_count', ''),
            'popularity': movie.get('popularity', ''),
            'production_companies': ', '.join([company['name'] for company in movie.get('production_companies', [])])
        }
        return movie_info
    else:
        print(f"Failed to retrieve details for movie ID {movie_id}. Status code: {response.status_code}")
        return None

# Function to get all movies and write them to a CSV
def get_all_movies():
    # Open the CSV file for writing
    with open('all_movie_details.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=[
            'movie_id', 'title', 'overview', 'genres', 'release_date', 'runtime', 'cast', 'crew', 
            'original_language', 'poster_path', 'vote_average', 'vote_count', 'popularity', 
            'production_companies', 'keywords'
        ])
        
        writer.writeheader()

        page = 1
        while True:
            # The TMDb API popular movie endpoint
            url = f'https://api.themoviedb.org/3/movie/popular?api_key={API_KEY}&language=en-US&page={page}'
            response = requests.get(url)

            if response.status_code == 200:
                data = response.json()
                if data['results']:
                    # Loop through the results and write each movie's details to the CSV file
                    for movie in data['results']:
                        movie_id = movie['id']
                        movie_info = get_movie_details(movie_id)

                        if movie_info:
                            writer.writerow(movie_info)
                            print(f"Written details for movie: {movie_info['title']}")

                    page += 1
                    time.sleep(1)  # To avoid hitting rate limits
                else:
                    # No more movies available (end of data)
                    break
            else:
                print(f"Failed to retrieve data on page {page}. Status code: {response.status_code}")
                break

    print(f"All movie details have been written to 'all_movie_details.csv'")

# Call the function to get all movies
get_all_movies()
```
